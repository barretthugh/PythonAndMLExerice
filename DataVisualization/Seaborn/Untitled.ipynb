```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'], kde=False, bins=30)
```

```python
sns.jointplot(x='total_bill',y='tip',
              data=tips, kind='hex')
```

```python
sns.pairplot(tips,hue='sex', palette='coolwarm')
```

```python
sns.rugplot(tips['total_bill'])
```

```python
import numpy as np
```

```python
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
```

```python
sns.countplot(x='sex', data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips, hue='sex', split=True)
```

```python
sns.stripplot(x='day',y='total_bill', data=tips,jitter=True,hue='sex',split=True)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='bar') # kind: kind of plot
```

```python
# Matrix Plots
```

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
tips.corr()
```

```python
sns.heatmap(tips.corr(),annot=True,cmap='coolwarm')
```

```python
fp = flights.pivot_table(index='month',columns='year', values='passengers')
```

```python
fp.head()
```

```python
sns.heatmap(fp,cmap='magma',linecolor='white',linewidths=1)
```

```python
sns.clustermap(fp,cmap='coolwarm',standard_scale=1)
```

```python
# Regression Plots
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,
           hue='sex',markers=['o','v'],scatter_kws={'s':100})
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,
           col='sex',row='time')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,
           col='day',row='time')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,
           col='day',hue='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,
           col='day',row='sex',aspect=0.6,size=8)
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

```python
sns.PairGrid(iris)
```

```python
import matplotlib.pyplot as plt
```

```python
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(plt.kdeplot) # seaborn and df version issue
```

```python
g = sns.FacetGrid(data=tips,col='time',row='smoker')
```

```python
g = sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(sns.distplot,'total_bill') # seaborn and df version issue
```

```python
g = sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(plt.scatter,'total_bill', 'tip')
```
