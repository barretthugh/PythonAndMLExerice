```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# FUNCTIONAL
plt.plot(x,y,'r*')
plt.xlabel('Xlabel')
plt.ylabel('Ylabel')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x, y, 'b')
plt.subplot(1,2,2)
plt.plot(y,x,'o')
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('lager plot')

axes2.plot(y,x)
axes2.set_title('small plot')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([1,1,1,1])
axes1.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
# axes.plot(x,y)
# plt.tight_layout()
# for current_axes in axes:
#     current_axes.plot(x,y)
axes[0].plot(x,y)
axes[0].set_title('first plot')

axes[1].plot(y,x)
axes[1].set_title('second plot')

plt.tight_layout()
```

```python
fig = plt.figure(figsize=(8,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))

# ax = fig.add_axes([0,0,1,1])
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig.savefig('my_picture.png',dpi=200)
```

```python
fig = plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y**2, label='y squared')
ax.plot(y,x**2, label='x squared')

ax.set_title('title')
ax.set_xlabel('Xlabel')
ax.set_ylabel('Ylabel')
ax.legend(loc=(0.1,0.1))
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='#ff8c00', lw=8, alpha=0.5,
        ls='--',markersize=30,marker='^',
       markerfacecolor='b',markeredgewidth=3,
       markeredgecolor='r')
ax.set_xlim([0,1])
ax.set_xlim([0,2])
```
